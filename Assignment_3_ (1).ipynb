Q1  What is the difference between a function and a method in Python ?

Answer -

 - Functions = Functions are nothing but a block of code which performs a specific task and can be called again and again as
per requirement in the code.

- Method = A method is a function defined inside a class and is associated with an object. It usually takes self as the first parameter to access the instance (object) it is being called on.



```python
# Function
def greet(name):
    return f"Hello, {name}!"

print(greet("Shivam"))
```

```python
# Method

class Person:
    def greet(self, name):
        return f"Hi, {name}! I'm a doctor"

p = Person()
print(p.greet("Shivam"))  # `greet` is a method
```

Q 6 . Explain the concept of generators in Python and how they are defined ?

Answer =

- Generator functions are a special type of function in Python that let you create iterators easily and save memory. Instead of returning all the data at once, they give you one value at a time — just when you need it.

- This is really helpful when you’re working with large datasets or long sequences, because it doesn’t load everything into memory at once.

```python
def countdown(n):
  while n > 0:
    yield n
    n -=1

countdown_gen = countdown(5)
for number in countdown_gen:
  print(number)
```

Q 7 What are the advantages of using generators over regular functions?

Answer =
- Use of yield in Python =
In a generator function, instead of using return, we use the yield keyword to send a value back to the caller.

- Lazy Evaluation in Generator = Functions
Generator functions follow the principle of lazy evaluation.

 # Why is it Useful?
 - Saves memory — doesn’t store the entire dataset in memory.
 - Speeds up performance when working with Large files
Big data Infinite sequences (like counters, streams)

- Memory Efficiency: Generators are memory-efficient because they don't store all values in memory at once.
Each value is generated and consumed one at a time, making them suitable for working with large datasets or
when memory usage is a concern.


```python
# use of yield

def pw_generator():
  yield 5
  yield 6
  yield "pw skills "

  # Using the generator
generator = pw_generator()
print(next(generator))
print(next(generator))
print(next(generator))

```

```python
# lazy evalution

def countdown(n):
  while n > 0:
    yield n
    n -=1

countdown_gen = countdown(5)
for number in countdown_gen:
  print(number)

```

```python
# memory efficiency
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Using the generator to print the first 10 Fibonacci numbers
fib_gen = fibonacci_generator()
for _ in range(10):
    print(next(fib_gen))
```

Q 10  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Answer =

1. map() Function:

- Applies a given function to each item in an iterable (like list, tuple, etc.).

  - Returns a map object (an iterator) which can be converted to a list, tuple, etc.

 - Syntax: map(function, iterable)

  - Commonly used for element-wise operations, like doubling values or converting strings to integers.

  - It does not modify the original iterable—returns a new one.
2. reduce() =

- reduce() Function (from functools module):
Applies a function cumulatively to the items of an iterable.

  - Reduces the iterable to a single value.

  - Syntax: reduce(function, iterable)

  - Works like: (((x1, x2), x3), x4)...

  - Used for aggregation tasks like sum, product, or finding max/min.

3. filter() =

- Applies a function that returns True or False to each item in an iterable.

  - Returns a filter object (an iterator) containing only the elements that satisfy the condition (i.e., where the function returns True).

 - Syntax: filter(function, iterable)

 - Commonly used to filter out elements based on a condition (e.g., keeping only even numbers).

 - Like map(), it does not modify the original iterable—returns a new filtered one.



```python
# map() function =

# calculate square of each number

numbers = [1, 5 , 7, 9, 25]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)
```

```python
# reduce function =

# find maximum no of list


from functools import reduce
numbers = [ 25 , 77 , 102 , 45 , 999]
max_number = reduce(lambda x, y: x if x > y else y,
numbers)
print(max_number)
```

```python
# filter function

nums = [1, 2, 3, 4, 5, 6]

# Filter even numbers
result = list(filter(lambda x: x % 2 == 0, nums))
print(result)
```

Q1 .  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list ?

ANSWER -

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example usage
my_list = [2, 3, 4, 6 , 22 , 44 , 39 , 27 ]
result = sum_even_numbers(my_list)
print("Sum of even numbers:", result)

```

Q 4  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n**0.5) + 1):  # check up to square root of n
        if n % i == 0:
            return False  # divisible by some number, not prime
    return True  # no divisors, so it's prime

# Example usage
for number in range(1, 201):
    if is_prime(number):
        print(number, "is a prime number")
```

Q 6 Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i
for power in powers_of_two(10):
    print(power, end=" ")
```

Q10 . Create a Python program that uses `filter()` to remove all the vowels from a given string.

Answer

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    result = ''.join(filter(lambda ch: ch not in vowels, input_str))
    return result

# Example usage
user_input = "Pharmacy is powerful"
filtered_string = remove_vowels(user_input)
print("String without vowels:", filtered_string)
```
